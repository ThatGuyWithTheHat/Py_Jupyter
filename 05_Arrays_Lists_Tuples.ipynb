```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.zeros(20)
x[:5] = 10
x[5:15] = np.arange(12, 31, 2)
x[15:] = 30
plt.plot(x)
plt.plot([4, 4], [8, 32],'k--')
plt.plot([14, 14], [8, 32],'k--')
plt.ylim(8, 32);
```

```python
alist = [1, 2, 3]
print('alist', alist)

btuple = (10, 20, 30)
print('btuple', btuple)

alist[0] = 7  # Since alist is a list, you can change values 
print('modified alist', alist)
print(2*alist)

#btuple[0] = 100  # TypeError: 'tuple' object does not support item assignment
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(0, 10, 0.2)
y = np.arange(0, 10, 0.2)
X, Y = np.meshgrid(x, y)
Z = np.array([[151, 176, 178],
       [121, 137, 130],
       [120, 125, 126]])
Z = X**2 + Y**2

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```

```python
import numpy as np 
x = np.array([[4, 2, 3, 2],
              [2, 4, 3, 1],
              [0, 4, 1, 3]])
print(x)
print(np.reshape(x, (6, 2)))  # 6 rows, 2 columns
print(np.reshape(x, (1, 12)))  # 1 row, 12 columns
```

```python
# arrays are not vectors or matrices (multiplication and division of two arrays is term by term)
import numpy as np 
a = np.arange(4, 20, 4)
b = np.array([2, 2, 4, 4])
print('array a:', a)
print('array b:', b)
print('a * b  :', a * b)  # term by term multiplication
print('a / b  :', a / b)  # term by term division
```

```python
x = np.array([[4, 2, 3, 2],
              [2, 4, 3, 1],
              [2, 4, 1, 3],
              [4, 1, 2, 3]])
print('the first row of x')
print(x[0])
print('the first column of x')
print(x[:, 0])
print('the third row of x')
print(x[2])
print('the last two columns of x')
print(x[:, -2:])
print('the four values in the upper right hand corner')
print(x[:2, 2:])
print('the four values at the center of x')
print(x[1:3, 1:3])
```

```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 20, 100)
y = np.sin(x)
plt.plot(x, y, 'b')
y[y > 0.5] = 0.5
y[y < -0.75] = -0.75
plt.plot(x, y, 'r');
```

```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
x = np.linspace(0, 6 * np.pi, 50)
y = np.sin(x)
plt.plot(x[y > 0], y[y > 0], 'bo')
plt.plot(x[y <= 0], y[y <= 0], 'ro');
```

```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
x = 10 - 20*np.random.random((500))
y = 10 - 20*np.random.random((500))
plt.plot(x, y, 'ro')
plt.plot(x[(x < -2) & (y >= -5) & (y < 0)], y[(x < -2) & (y >= -5) & (y < 0)], 'bo')
r = np.sqrt((x - 5) ** 2 + y ** 2)
plt.plot(x[r < 5], y[r < 5], 'go')
plt.axis('scaled');
```
